```python
import pandas as pd

#input_data = input("Introduce path for data")
data = pd.read_csv("data_processing/company_bankruptcy_clean.csv").drop(["Bankrupt?"], axis=1)#(input_data)

#Para prueba:
data_sample = data.head(20)
data_dict = data_sample.to_dict(orient='list')

data
```

```python
import json

with open("uri2.json", "r") as suri:
    scoring_uri = json.load(suri)["URI"][0]
```

```python
#Formateado para la API:
data_json = json.dumps({"data": [data_dict]})

suri = open("uri2.json", "r")
scoring_uri = json.load(suri)["URI"][0]
suri.close()
```

```python
print("URI:", scoring_uri[:100])  #Para asegurar que s√≠ se carga la URI de forma correcta
```

```python
## La API per-s√©:

headers = {"Content-Type": "application/json"}
import requests
response = requests.post(scoring_uri, data=data_json, headers=headers)

try:
  if response.status_code == 200:
    print (f"[SUCCESS]:")
    result = json.loads(response.json())
    print(result)
    data["Exited"] = result
    display(data)
  else:
    print (f"[CODE ERROR]:")
    print(f"Error: {response.text}")
except OSError as e:
  print ("OSError: ", e)
  if e.errno == 36:
    print("Invalid argument error: ", e)
```
